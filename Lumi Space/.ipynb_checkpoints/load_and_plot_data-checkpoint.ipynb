## Accessing the data

```python
import numpy as np
import json
import glob
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Location of the data on disk
data_path = "data/"

# Print the data files 
print(glob.glob(data_path + "*.json"))
```

## Description of orbital data
You can see that there are four files present above: two for each satellite Envisat and globalstar. The files that end in "state.json" contain orbital data: satellite positions, velocities and also attitude (see e.g. https://www.aero.iitb.ac.in/satelliteWiki/index.php/What_is_Attitude%3F) in terms of pitch, roll and yaw. All data in this file is in an Earth centred inertial coordinate system, but don't worry too much about the details. Let's load the orbital data first...

## Load orbital data for Envisat
This cell will load data for the now defunct Envisat satellite (in the image below) but Globalstar data can also be loaded by using the appropriate path above.

<div>

</div>


```python
# path to Envisat orbital data.
envisat_orbit_path = 'data/envisat_state.json'

# load the data from file into orbit dictionary
with open(envisat_orbit_path, 'r') as f:
    orbit = json.load(f)

# Convert all orbital data to numpy arrays for easier plotting.
for key in orbit.keys():
    orbit[key] = np.array(orbit[key])

# What data do we have?
print("Data available:\n", orbit.keys())    
```

## Plotting orbital data
Lets plot some satellite positions to see their orbits. Position data has an array shape of (time, 3) where time gives the position at a specific time and 3 represent the x,y,z cartesian coordinates. Positions are in units of m and velocities are in units of m/s.

```python
pos = orbit['pos']

fig = plt.figure()
# Plot a sphere to represent the Earth.
ax = fig.add_subplot(111, projection='3d')
r = 6000000
u, v = np.mgrid[0:2 * np.pi:30j, 0:np.pi:20j]
x = r*(np.cos(u) * np.sin(v))
y = r*(np.sin(u) * np.sin(v))
z = r*(np.cos(v))
ax.plot_surface(x, y, z, cmap=plt.cm.YlGnBu_r)

# Plot the satellite orbit.
x = pos[:, 0]
y = pos[:, 1]
z = pos[:, 2]
ax.plot(x, y, z)
ax.view_init(10, 75)
plt.show()
```

## Decription of groundstation pass data
The files that end in "passes.json" contain ground station pass data. This is data about the passage of Envisat and Globalstar as seen from two groundstations: the first station is Graz, Austria and the second is at Herstmonceux, UK. Use the keys defined here (graz_idx and herstmonceux_idx) throughout to access data for a given groundstation.

```python
# path to Envisat orbital data.
envisat_passes_path = 'data/envisat_passes.json'

graz_idx=0
herstmonceux_idx=1

# load the data from file into groundstations dictionary
with open(envisat_passes_path, 'r') as f:
    groundstations = json.load(f)
    
# convert to numpy arrays for easier plotting    
for groundstation in groundstations:
    for gs_pass in groundstation['passes']:
        for key in gs_pass.keys():
            if type(gs_pass[key]):
                gs_pass[key] = np.array(gs_pass[key])    

# print out the keys present in the passes data
print(groundstations[graz_idx].keys())
print(groundstations[herstmonceux_idx].keys())

# print the latitude and longitude of the two groundstations
print("Graz. Latitude:", groundstations[graz_idx]['lat'], ". Longitude:", groundstations[graz_idx]['lon'])
print("Herstmonceux. Latitude:", groundstations[herstmonceux_idx]['lat'], ". Longitude:", groundstations[herstmonceux_idx]['lon'])
```

### Accessing pass data
A list of passes above each groundstation can be found under the 'passes' dictionary element. There are ~6 passes over a 24 hour period for each groundstation and these can be accessed with pass_idx. Note: pass_idx=0 contains an empty pass and should be ignored.

```python
pass_idx=1
# print out the data present from a groundstation pass
print(groundstations[graz_idx]['passes'][pass_idx].keys())

# print out the time of the groundstation pass
print("start time of pass (in units of seconds since 00:00:00 on 1/1/2000):", groundstations[graz_idx]['passes'][pass_idx]['time'][0])
```

Other data here that might be interesting to consider are:
- 'w': the satellite rotational (not orbital) angular velocity.
- 'gs vec': a vector in the interial coordinate system that represents the location of the groundstation.
- 'pos': a vector in the interial coordinate system that represents the location of the satellite.
- 'vel': a vector in the interial coordinate system that represents the velocity of the satellite.
- 'range': the range between the satellite and the groundstation as measured by Satellite Laser Ranging (SLR).

# Plotting ground station pass data
Let's plot some groundstation pass data. This polar azimuth/elevation plot generated by the cell below shows the passage of a satellite overhead at the Graz groundstation. The azimuth is measured from 0-360 from North and the elevation is measured in degrees above the local horizon at the groundstation. For more information on this coordinate system see: https://en.wikipedia.org/wiki/Local_tangent_plane_coordinates.



```python
fig = plt.figure(dpi=300)
ax1 = fig.add_subplot(221, projection='polar')
pass_idx=2
azimuth = groundstations[graz_idx]['passes'][pass_idx]["azimuth"]
elevation = groundstations[graz_idx]['passes'][pass_idx]["elevation"]
ax1.plot(azimuth, elevation*(180.0/np.pi))
ax1.set_rlim(bottom=90, top=0)
ax1.set_rticks([90,60,30,0])
ax1.set_rlabel_position(-22.5) 
ax1.grid(True)
```

## Plotting attitude data
Next, lets plot some data that desscribes the attitude of the satellite in orbit as it follows the azimuth and elevation in the plot above. For more information on yaw/pitch/roll check out this wiki page: https://en.wikipedia.org/wiki/Aircraft_principal_axes

```python
plt.figure(dpi=300)
pitch = groundstations[graz_idx]['passes'][pass_idx]['pitch']
roll = groundstations[graz_idx]['passes'][pass_idx]['roll']
yaw = groundstations[graz_idx]['passes'][pass_idx]['yaw']
time = groundstations[graz_idx]['passes'][pass_idx]['time']
time = time-time[0]
plt.plot(time, pitch, label="pitch")
plt.plot(time, roll, label="roll")
plt.plot(time, yaw, label="yaw")
plt.grid()
plt.legend()
plt.xlabel("elapsed time (seconds)")
plt.ylabel("angle (radians)")
plt.show()
```

Note that attitude data such as this is also available for the whole orbit rather than just for a groundstation pass in the orbital data dictionary.
